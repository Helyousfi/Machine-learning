# Matplotlib example workflow

```python
%matplotlib inline
# 0. import matplotlib
import matplotlib.pyplot as plt

# 1. prepare the data
x = [1, 2, 3, 4]
y = [2, 5, 3, 4]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10))

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title="Simple Plot",
        xlabel="x-axis",
        ylabel="y-axis")

# 5. save the fig
fig.savefig("sample-plot.png")
```

```python
car_sales = pd.read_csv("7.1 car-sales.csv")
car_sales.head()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# plot total sales
car_sales["Total Sales"] = car_sales["Total Sales"].astype(int)
fig, ax = plt.subplots()
car_sales.plot(figsize=(10, 5), x="Sale Date", y="Total Sales", kind="bar", ax=ax);
fig.savefig('i1.jpg')
```

```python
heart_disease = pd.read_csv("13.1 heart-disease.csv")
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 40]
over_50.head()
```

```python
over_50.plot(kind = "scatter",
            x = 'age',
            y = 'chol',
            c = 'target',
            figsize = (6, 6));
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);
ax.set_xlim([45, 80]);
fig.savefig('blog/fig.jpg')
```

```python
fig, ax  = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target']);
ax.set(title="heart disease and cholesterol levels",
    xlabel = "age",
    ylabel="cholesterol");
ax.legend(*scatter.legend_elements(), title="Target");
fig.savefig('blog/fig.jpg')
```

#### Small test

```python
import numpy as np
data = (np.random.rand(1000, 4) * 100).astype(int)
test_dataframe = pd.DataFrame(data=data, columns=['a', 'b', 'c', 'd'])
test_dataframe["label"] = (test_dataframe["d"] > 50).astype(int)

fig, ax = plt.subplots(figsize = (10, 6))
scatter = ax.scatter(x = test_dataframe['a'],
                    y = test_dataframe['b'], 
                    c = test_dataframe["label"])
ax.set(title = "Random samples with random labels", xlabel="x axis", ylabel = "y axis")
ax.legend(*scatter.legend_elements(), title="legend")
ax.axhline(test_dataframe["d"].mean(), linestyle="--");
```
