```python
import time
def plot_band(dataset):
    for i in range(0, 200, 50):
        plt.figure(figsize=(16, 12))
        band_no = i
        plt.imshow(dataset[:,:, band_no], cmap='jet')
        plt.title(f'Band-{band_no}', fontsize=14)
        plt.axis('on')
        plt.colorbar()
        plt.show()
plot_band(X)
```
